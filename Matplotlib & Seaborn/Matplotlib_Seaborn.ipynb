# Bar chart

### Using Matplotlib

```python
#importing matplotlib & pandas
import matplotlib.pyplot as plt

import pandas as pd

data = {
    "Grades" : ["A", "A-", "B+", "B"],
    "Ratios" : [3, 5, 20, 12]
}

df = pd.DataFrame(data)

#Using the barchart

plt.bar(df["Grades"], df["Ratios"])

#Labeling
plt.xlabel("Grades")
plt.ylabel("Ratios")

#Title
plt.title("Fig- NSU section")
plt.grid()

#Show
plt.show()
```

### Using Seaborn

```python
#importing matplotlib & seaborn & pandas
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = {
    "Grades" : ["A", "A-", "B+", "B"],
    "Ratios" : [3, 5, 20, 12]
}

df = pd.DataFrame(data)

plt.figure(figsize = (12,8))

#Using the barchart

sns.barplot(x = df["Grades"], y = df["Ratios"])

#Labeling
plt.xlabel("Grades")
plt.ylabel("Ratios")

#Title
plt.title("Fig- NSU section")
plt.grid()

#Show
plt.show()
```

# Choosing Seaborn Color Palettes

Seaborn https://seaborn.pydata.org/tutorial/color_palettes.html

Matplotlib https://matplotlib.org/stable/gallery/color/named_colors.html

```python
#importing

import pandas as pd
import matplotlib.pyplot as pt
import seaborn as sb

# Creating data

shantir_neer = {
    "Labours" : ["Elias", "Khokon", "Shamim", "Liakat", "Shopon", "Salman"],
    "Salaries" : [40, 11, 10, 5, 30, 15]
}

df = pd.DataFrame(shantir_neer)

#List of color palletes
palette_list = ["deep", "muted", "bright", "pastel", "dark", "colorblind"]

#Set size
pt.figure(figsize = (12, 8))

#Sub Plotting usnig seaborn

for i, pl in enumerate(palette_list):
    pt.subplot(3,3, i+1) # Set size
    sb.barplot(data = df, x = df["Labours"], y = df["Salaries"], palette = pl) # customizing using seaborn
    pt.grid()
    pt.title("nsu.siam")

#Applying tight layout
pt.tight_layout()

#Show
pt.show()
```

# Function to Save Figures

```python
# code to save the figures as high-res PNGs
import os
from pathlib import Path
from matplotlib import pyplot as plt

IMAGES_PATH = Path() / "Images/Data_Visualization"
IMAGES_PATH.mkdir(parents=True, exist_ok=True)

def save_fig(fig_id, tight_layout=True, fig_extension="png", resolution=300):
    path = IMAGES_PATH / f"{fig_id}.{fig_extension}"
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)
```

# Count Plot

 A count plot, in the context of data visualization using libraries like Seaborn or Matplotlib, is a type of categorical plot that displays the count of observations in each category using bars.

The count plot is particularly useful when dealing with categorical data, where you want to visualize the frequency or distribution of observations within different categories. It represents the number of occurrences of each category in the dataset using vertical bars.

https://seaborn.pydata.org/generated/seaborn.countplot.html

```python
# importing libraries

import pandas as pd
import matplotlib.pyplot as pt
import seaborn as sb

# Custom dictionary
data = {
    "allergies" : ['Yes', 'No', "Not Sure", "Yes", "No", "Not Sure", 'Yes', 'No','Maybe', 'Yes']
}

df = pd.DataFrame(data)

pt.figure(figsize = (12, 8))

# Creating count plot
sb.countplot(data = df, x = df["allergies"])

pt.title("Allergy count")
pt.grid()

pt.xlabel("Is allergy")
pt.ylabel("Count")

#Save the pic

save_fig("Allergy Value Count")

#Show

pt.show()
```

# Line plot
 Used for see the changes in a time interval on the continuous data

```python
# Create a dictionary for the dataset

import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Date" : pd.date_range("2022-01-01", periods = 365),
    "Values" : np.random.uniform(low = 0, high = 100, size = 365)
})

#Creating Line plot

import matplotlib.pyplot as pt
import seaborn as sb

pt.figure(figsize = (12, 8))

sb.lineplot(data = df, x = "Date", y = "Values", color = "skyblue")

#Customizing

pt.title("Line Plot - Date vs Values")
pt.xlabel("Date")
pt.ylabel("Values")

# Rotating xlabel

plt.xticks(rotation=45)

#saving the figure

save_fig("Line Plot - Date vs Values")

#showing

pt.show()
```

# Boxplot

* **Key Features of a Box Plot**:

***Five Summary Statistics***: A box plot represents the following statistical measures:

**Minimum**: The smallest data point within the dataset excluding outliers.

**First Quartile (Q1)**: The value below which 25% of the data falls (25th percentile).

**Median (Q2)**: The middle value of the dataset (50th percentile or second quartile).

**Third Quartile (Q3)**: The value below which 75% of the data falls (75th percentile).

**Maximum**: The largest data point within the dataset excluding outliers.



```python
# Boxplot

sb.boxplot(x = df["bmi"])

plt.xlabel('BMI')
plt.title("Box Plot")
plt.grid()
save_fig("Box Plot")
plt.show()
```

```python
# Analysis

big_data = df["bmi"]
Q1 = big_data.quantile(0.25)
Q2 = big_data.median()
Q3 = big_data.quantile(0.75)
IQR = Q3 - Q1

minimum = big_data.min()
maximum = big_data.max()

# Print

print("Q1:", Q1)
print("Mean:", Q2)
print("Q3:", Q3)
print("IQR:", IQR)
print("Minimum:", minimum)
print("Maximum:", maximum)

df.describe()
```

# Pair plot

```python
# Creating a dictionary for creating a custom dataset

import numpy as np
import pandas as pd
import matplotlib.pyplot as pt
import seaborn as sb

df = pd.DataFrame(
    {
        "A" : np.random.randn(100),
        "B" : np.random.uniform(low = 0, high = 100, size = 100),
        "C" : np.random.normal(loc = 2, scale = 1, size = 100)
    }
)

#Creating pair plot

pt.figure(figsize = (8, 6))

sb.pairplot(df)

#Customizing

pt.suptitle("Pairplot", y=1.02)

#saving the figure

save_fig("Pairplot")

#showing

pt.show()
```
