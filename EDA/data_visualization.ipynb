# Data Visualization

```python
# Load libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as pt
import seaborn as sb
```

```python
# Load the dataset
df = pd.read_csv("diabetes_binary_5050split_health_indicators_BRFSS2015 - diabetes_binary_5050split_health_indicators_BRFSS2015.csv")
```

```python
# code to save the figures as high-res PNGs
import os
from pathlib import Path
from matplotlib import pyplot as plt

IMAGES_PATH = Path() / "Images/Data_Visualization"
IMAGES_PATH.mkdir(parents=True, exist_ok=True)

def save_fig(fig_id, tight_layout=True, fig_extension="png", resolution=300):
    path = IMAGES_PATH / f"{fig_id}.{fig_extension}"
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)
```

## Diabetes target variable count

```python
# set the size for plotting
pt.figure(figsize = (7,5))

# creating a pie chart
pt.pie([df["Diabetes_binary"].value_counts()[0], df["Diabetes_binary"].value_counts()[1]], 
       labels = ['No Diabetes (0)', 'Diabetes (1)'], autopct = "%1.1f%%", startangle = 90)

# Customizing the plot
pt.title("Diabetes Records")

# saving the img
save_fig("Distribution of Diabetes Records")

# showing the plot
pt.show()
```

## Diabetes Records per Gender

```python
# 0 = no diabetes
# 1 = pre-diabetes
# 0 - female
# 1 - male

# Create a count plot
sb.countplot(data = df, x = df["Diabetes_binary"], hue = "Sex", palette = "rocket")

# Customizing
pt.xlabel("Diabetes status", size = 10, fontstyle = "oblique", weight = 650)
pt.ylabel("Count", size = 10, fontstyle = "oblique", weight = 650)
pt.grid()
pt.title("Diabetes Records - Genderwise")

# Customization for differentiate
pt.legend(title = "Gender", labels = ["Female", "Male"])
pt.xticks(ticks = [0,1], labels = ["No Diabetes", "Pre Diabetes"])

# Saving the figure
save_fig("Diabetes Records per Gender")

# Showing
pt.show()
```

# Relationship between diabetes and stroke

```python
# 0 = no stroke
# 1 = stroke
# 0 - no diabetes
# 1 - pre diabetes

# Create a count plot
sb.countplot(data = df, x = df["Diabetes_binary"], hue = "Stroke", palette = "viridis")

# Customizing
pt.xlabel("Diabetes status", size = 10, fontstyle = "oblique", weight = 650)
pt.ylabel("Count", size = 10, fontstyle = "oblique", weight = 650)
pt.grid()
pt.title("Diabetes Records - Strokewise")

# Customization for differentiate
pt.legend(title = "Stroke status", labels = ["NO", "YES"])
pt.xticks(ticks = [0,1], labels = ["No Diabetes", "Pre Diabetes"])

# Saving the figure
save_fig("Diabetes Records per Stroke occurrence")

# Showing
pt.show()
```

## Relationship between diabetes and Blood Pressure

```python
# Create a count plot
sb.countplot(data = df, x = df["Diabetes_binary"], hue = "HighBP", palette = "mako")

# Customizing
pt.xlabel("Diabetes status", size = 10, fontstyle = "oblique", weight = 650)
pt.ylabel("Count", size = 10, fontstyle = "oblique", weight = 650)
pt.grid()
pt.title("Diabetes Records - BPwise")

# Customization for differentiate
pt.legend(title = "Blood Pressure Level", labels = ["Low", "High"])
pt.xticks(ticks = [0,1], labels = ["No Diabetes", "Pre Diabetes"])

# Saving the figure
save_fig("Diabetes Records per Blood pressure lavel")

# Showing
pt.show()
```

## Histogram of BMI

```python
# Creating a histplot
sb.histplot(data = df, bins = 20, x = df["BMI"], kde = True, color = "brown")

# Customizing
pt.xlabel("BMI")
pt.ylabel("Frequency")
pt.grid()
pt.title("Histplot of BMI")

# saving the figure
save_fig("BMI - Histogram analysis")

# showing
pt.show()
```
