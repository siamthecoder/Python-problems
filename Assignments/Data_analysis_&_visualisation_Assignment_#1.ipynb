```python
## For creating a dataset from the informations in given csv file

import pandas as pd

df = pd.read_csv("Life_Expectancy_Data.csv")

df.info()
```

# Function to Save Figures

```python
# code to save the figures as high-res PNGs
import os
from pathlib import Path
from matplotlib import pyplot as plt

IMAGES_PATH = Path() / "Images/Data_Visualization"
IMAGES_PATH.mkdir(parents=True, exist_ok=True)

def save_fig(fig_id, tight_layout=True, fig_extension="png", resolution=300):
    path = IMAGES_PATH / f"{fig_id}.{fig_extension}"
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)
```

## Question 01.

* üë®‚Äçüíª***Analyse the avg life expectancy rate in Argentina & present the countries population growth over the years.***


# Ans.to.the.Ques.No.01:

```python
# filtering the informations of Argentina from the created dataset
data_argentina = df.query('Country.str.startswith("Argentina")')

data_argentina
```

```python
# For calculating the avg life expectency in Argentina
import numpy as np
avg_arg_life = np.mean(data_argentina["Life_expectancy"])

print("Average life expectency in Argentina is: ",avg_arg_life,"years")
```

```python
data_argentina.info()
```

```python
# Presenting the population growth over the years in Argentina using Line plot.
import matplotlib.pyplot as pt
import seaborn as sb

# Set the size for plotting
pt.figure(figsize = (12, 8))

# Using seaborn for line plot
sb.lineplot(data = data_argentina, x = df["Year"], y = df["Population_mln"], color = "blue", n_boot = 1)

# Customizing the plot
pt.title("Argentina - Population growth over the years")
pt.xlabel("Years")
pt.ylabel("Population(mln)")
pt.grid()

# rotating the xlabel
pt.xticks(rotation = 45)

# saving the figure
save_fig("Argentina - Population growth over the years")

# showing
pt.show()
```

## üìäConclusion

* ***So, we can state from our above analysis that; in Argentina the average life expectency rate is 74.8625 years. In addition, from our line plot we can state that over the years the population growth is increasing in Argentina. Where, the lowest number of population was in 2004 & the highest number of population was in 2012.***

# Question 02.

* üë®‚Äçüíª***Find the number of developed countires then visualize it for analysing the difference of developement.***


# Ans.to.the.Ques.No.02 :

```python
# Cleaning the data by droping the duplicates for getting the main list of countires

cleaned_country_list = df.drop_duplicates(subset = "Country", keep = "first")
```

```python
cleaned_country_list.info()
```

```python
# Finding the number of developed countries [0 : Not developed; 1: Developed]
cleaned_country_list["Economy_status_Developed"].value_counts()
```

```python
# Now visualizing our result using countplot
import matplotlib.pyplot as pt
import seaborn as sb

# Set the size for plotting
pt.figure(figsize = (8, 4))

# Using seaborn for countplot
sb.countplot(data = cleaned_country_list, x = cleaned_country_list["Economy_status_Developed"], palette = "dark")

# Customizing the plot
pt.grid()
pt.xlabel("Developed Status", size = 8, fontstyle = "italic", weight = 500)
pt.ylabel("Country Count", size = 8, fontstyle = "italic", weight = 500)
pt.title("Number of Developed countries")

# saving the figure
save_fig("Number of Developed countries")

# showing
pt.show()
```

## üìäConclusion

* ***So, we can state from our above analysis that; in our dataset there are only 37 developed countries among 179 taken countries. Also from our countplot it is clearly visible that development differences are high in our dataset.***

# Question 05 .
* üë®‚Äçüíª***Analyse the BMI in the dataset then analyse & visualize the BMI for the year 2015.***


# Ans.to.the.Ques.No.05 :

```python
# analyzing the BMI by calculations
import numpy as np
bmi_data = df["BMI"]

mean_bmi = np.mean(bmi_data)


minimum_bmi = bmi_data.min()
maximum_bmi = bmi_data.max()

print("Average BMI: ", mean_bmi)
print("Minimum BMI: ", minimum_bmi)
print("Maximum BMI: ", maximum_bmi)
```

```python
# filtering the datas for year 2015

df_2015 = df.query('Year == 2015')
```

```python
# Using Box plot to visualize the BMI for the year 2015
import matplotlib.pyplot as pt
import seaborn as sb

# Set size
pt.figure(figsize = (6, 4))

# Using Box plot
sb.boxplot(x = df_2015["BMI"])

# Customizing the plottings
pt.xlabel("BMI")
pt.title("Box Plot for BMI in 2015")
pt.grid()

# Saving the figure
save_fig("Box Plot for BMI in 2015")

# showing
plt.show()
```

```python
# Analysis from Box plot

analysing_2015_bmi = df_2015["BMI"]

Q1 = analysing_2015_bmi.quantile(0.25)
Q2 = analysing_2015_bmi.median()
Q3 = analysing_2015_bmi.quantile(0.75)
IQR = Q3 - Q1

minimum = analysing_2015_bmi.min()
maximum = analysing_2015_bmi.max()

# Printing the Q1, Q2. Q3, IQR, minimum, maximum

print("Q1: ", Q1)
print("Median: ", Q2)
print("Q3: ", Q3)
print("IQR: ", IQR)
print("Minimum: ", minimum)
print("Maximum: ", maximum)

df_2015.describe()
```

## üìäConclusion

* *for overall dataset---> Average BMI: 25.03, Minimum BMI: 19.8, Maximum BMI: 32.1*


* *for the year 2015---> The value below which 25% of the data falls:  23.8, The middle value of the dataset: 26.2, The value below which 75% of the data falls: 27.0, Maximum BMI: 20.5 & Minimum BMI: 32.1 which is also visualized by our box plot*

# Question 06.
* üë®‚Äçüíª***Analyse the Regional HIV cases.***


# Ans.to.the.Ques.No.06 :

```python
# Using Swarm Plot for analysing the regional HIV case
import matplotlib.pyplot as pt
import seaborn as sb

# Set the size
pt.figure(figsize = (12, 8))

# Creating swarm plot
sb.swarmplot(data = df, x = df["Region"], y = df["Incidents_HIV"], color = "green", size = 3)

# Customizing
pt.title("Regional HIV Case Ratio")
pt.xlabel("Regions")
pt.ylabel("HIV %")
pt.grid()
pt.tight_layout()

pt.xticks(rotation = 45)

# save the figure
save_fig("Regional HIV Case Ratio")

# show the figure
pt.show()
```

## üìäConclusion

* ***So from this Swarm Plot we can easily represent that; in Africa region the HIV case ratio is highest among other 8 regions***
